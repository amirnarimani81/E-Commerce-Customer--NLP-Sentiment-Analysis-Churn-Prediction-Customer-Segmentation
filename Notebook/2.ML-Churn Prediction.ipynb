<p style="text-align: justify; font-family: cursive; font-size: 14px; color: green; font-weight: bold">
<b>🔹 ROC Curve (Receiver Operating Characteristic)</b><br>
ROC measures the model’s ability to rank positive cases higher than negative cases. A high AUC indicates the model can effectively distinguish between churners and non-churners. ROC is best used for overall model comparison.
</p>

<p style="text-align: justify; font-family: cursive; font-size: 14px; color: green; font-weight: bold">
<b>🔹 Precision–Recall Curve</b><br>
PR curves are more informative when the dataset is imbalanced. Precision shows how many predicted churners are correct, while recall shows how many actual churners were detected. High precision means fewer false churn alerts; high recall means fewer missed churners.
</p>

<p style="text-align: justify; font-family: cursive; font-size: 14px; color: green; font-weight: bold">
<b>✅ Best Practice Tip</b><br>
If churn rate is below 30% (common in telecom), PR curves are usually more meaningful than ROC because they focus on the minority class (churners).
</p>

<p style="text-align: justify; font-family: cursive; font-size: 14px; color: green; font-weight: bold">
<b>🔍 Quick Definitions</b><br>
Precision = TP / (TP + FP): Out of all predicted churners, how many actually churned?<br>
Recall = TP / (TP + FN): Out of all actual churners, how many did we correctly identify?
</p>

<p style="text-align: justify; font-family: cursive; font-size: 14px; color: green; font-weight: bold">
<b>⚖️ Threshold Tuning</b><br>
Changing the decision threshold lets you balance Precision and Recall based on business goals:
</p>

<p style="text-align: justify; font-family: cursive; font-size: 14px; color: green; font-weight: bold">
<b>🏥 Disease Screening:</b> maximize recall by lowering the threshold to catch more positives.<br>
<b>🛡️ Spam Detection:</b> maximize precision by raising the threshold to reduce false positives.<br>
<b>🔍 Fraud Detection:</b> balance precision and recall using F1-score or a custom cost-based metric.
</p>

<p style="text-align: justify; font-family: cursive; font-size: 14px; color: green; font-weight: bold">
<b>🔹 Precision</b><br>
Shows the accuracy of positive predictions. Higher precision means fewer false positives.
</p>

<p style="text-align: justify; font-family: cursive; font-size: 14px; color: green; font-weight: bold">
<b>🔹 Recall</b><br>
Shows how many actual positives are detected. Higher recall means fewer false negatives.
</p>

<p style="text-align: justify; font-family: cursive; font-size: 14px; color: green; font-weight: bold">
<b>🔹 F1-Score</b><br>
The harmonic mean of precision and recall. It is a balanced metric for imbalanced datasets.
</p>

</div>


```python
lr_clf = LogisticRegression(solver='saga', penalty='elasticnet', class_weight='balanced', C=1.0, l1_ratio=0.5, max_iter=10000)


    #penalty='l2',                # Regularization type (L1 or L2)
    #C=1.0,                       # Regularization strength (higher C means less regularization)
    #max_iter=100,                # Maximum number of iterations for optimization
    #multi_class='auto',          # For multi-class classification (binary is default)
   # random_state=42,             # Random state for reproducibility
    #fit_intercept=True,          # Whether to fit the intercept term or not
    #class_weight=None,           # To handle class imbalance (set 'balanced' to adjust weights)
    #warm_start=False,            # Whether to reuse the solution of the previous call to fit (for multiple fitting calls)
    #n_jobs=None,                 # Number of CPU cores to use during computation (None means 1 core)




lr_clf.fit(X_train_smote, y_train_smote)
```

<div style="border-radius: 5px; -webkit-border-radius: 5px; -moz-border-radius: 5px; font-family: cursive; border: 3px solid #008000; text-align: justify; color: black; font-size: 14px; padding: 5px; background:#F2FFFF">
  <p style="text-align: justify; font-family: cursive; font-size: 14px; color: green; font-weight: bold">
    🧪 <b>What is an ROC Curve?</b>
  </p>
  <p>
    The <b>Receiver Operating Characteristic (ROC) curve</b> visualizes how well a classifier distinguishes between classes. It plots:
  </p>
  <ul>
    <li><b>True Positive Rate (TPR)</b> = Sensitivity = Recall = TP / (TP + FN)  
        → Measures how many actual positives the model correctly identifies.</li>
    <li><b>False Positive Rate (FPR)</b> = FP / (FP + TN)  
        → Measures how many actual negatives are incorrectly predicted as positive.</li>
  </ul>
  <p>
    Each point on the ROC curve corresponds to a different <b>decision threshold</b>. Changing the threshold shifts the balance between TPR and FPR.
  </p>

  <p style="text-align: justify; font-family: cursive; font-size: 14px; color: green; font-weight: bold">
    🧮 <b>What is AUC?</b>
  </p>
  <p>

  <p style="text-align: justify; font-family: cursive; font-size: 14px; color : green ; font-weight : bold">
    K-Nearest Neighbors (KNN) Hyperparameter Explanation
  </p>

  <b>🔸 n_neighbors = K</b><br>
  This parameter defines the number of nearest data points (<b>K</b>) used to make a prediction.

  <br><br>
  <b>Example:</b>
  <ul>
    <li>If <b>K = 5</b>, the algorithm identifies the <b>5 closest neighbors</b> to the query point.</li>
    <li>The predicted class is determined by aggregating the labels of these neighbors.</li>
  </ul>

  <b>Impact:</b>
  <ul>
    <li>Small K → sensitive to noise (high variance).</li>
    <li>Large K → smoother decision boundary (higher bias).</li>
  </ul>

  <br>
  <b>🔸 weights = 'distance'</b><br>
  This parameter controls how much influence each neighbor has on the prediction.

  <ul>
    <li><b>'uniform'</b> (default): All neighbors contribute equally.</li>
    <li><b>'distance'</b>: Closer neighbors have a stronger influence than farther ones.</li>
  </ul>

  <b>Interpretation:</b><br>
  When using distance-based weighting, a neighbor at distance <b>0.5</b> contributes more to the
  prediction than a neighbor at distance <b>1.5</b>. This is particularly useful when nearby
  points are more reliable or informative.

  <br><br>
  <b>🔸 p (Minkowski Distance Parameter)</b><br>
  This parameter defines the distance metric used to measure similarity between data points.

  <br><br>
  <b>p = 1 → Manhattan Distance</b>
  <br>
  <i>Grid-based distance (horizontal + vertical movement)</i>
  <br>
  d = |x<sub>1</sub> − x<sub>2</sub>| + |y<sub>1</sub> − y<sub>2</sub>|

  <br><br>
  <b>p = 2 → Euclidean Distance (Default)</b>
  <br>
  <i>Straight-line distance between points</i>
  <br>
  d = √[(x<sub>1</sub> − x<sub>2</sub>)² + (y<sub>1</sub> − y<sub>2</sub>)²]

  <br><br>
  <b>Conclusion:</b><br>
  Setting <b>p = 2</b> means the KNN model uses <b>Euclidean distance</b> to evaluate closeness,
  which is the most common and intuitive choice for continuous numerical features.
</div>

```python
test_error_rates = []
train_error_rates = []


for k in range(1,10):
    KNC1 = KNeighborsClassifier(n_neighbors=k,
        weights="distance",         # optional but recommended
        metric="minkowski")
    KNC1.fit(X_train_smote, y_train_smote)

    scores = cross_validate(KNC1, X_train_smote, y_train_smote, scoring = ['accuracy'], cv = 10, return_train_score=True)

    accuracy_test_mean = scores["test_accuracy"].mean()
    accuracy_train_mean = scores["train_accuracy"].mean()


    test_error = 1 - accuracy_test_mean
    train_error = 1 - accuracy_train_mean
    test_error_rates.append(test_error)
    train_error_rates.append(train_error)

```

```python
results_svm = {'fold': [], 'accuracy': [], 'sensitivity': [], 'specificity': [], 'f1_score': [], 'roc_auc': []}

for fold in range(2, 11):
    kf = KFold(n_splits=fold, shuffle=True, random_state=42)

    acc_scores = []
    sens_scores = []
    spec_scores = []
    f1_scores = []
    roc_scores = []

    for train_index, test_index in kf.split(X, y):
        X_tr, X_te = X.iloc[train_index], X.iloc[test_index]
        y_tr, y_te = y.iloc[train_index], y.iloc[test_index]

        # Fit inside loop (important!)
        best_model_svm.fit(X_tr, y_tr)

        y_pred = best_model_svm.predict(X_te)
        y_proba = best_model_svm.predict_proba(X_te)[:, 1]

        acc_scores.append(accuracy_score(y_te, y_pred))
        sens_scores.append(recall_score(y_te, y_pred, average='weighted'))

        tn, fp, fn, tp = confusion_matrix(y_te, y_pred).ravel()
        spec_scores.append(tn / (tn + fp))

        f1_scores.append(f1_score(y_te, y_pred, average='weighted'))
        roc_scores.append(roc_auc_score(y_te, y_proba))

    results_svm['fold'].append(fold)
    results_svm['accuracy'].append(np.mean(acc_scores))
    results_svm['sensitivity'].append(np.mean(sens_scores))
    results_svm['specificity'].append(np.mean(spec_scores))
    results_svm['f1_score'].append(np.mean(f1_scores))
    results_svm['roc_auc'].append(np.mean(roc_scores))

results_df_svm = pd.DataFrame(results_svm)
results_df_svm
```

```python

metrics = []

for name, model in models.items():

    # Train
    model.fit(X_train_smote, y_train_smote)
    
    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Probabilities / scores for ROC-AUC
    if hasattr(model, "predict_proba"):
        y_test_proba = model.predict_proba(X_test)
        # For binary classification take class 1
        if y_test_proba.shape[1] == 2:
            y_test_proba = y_test_proba[:, 1]
    else:  # e.g., SVM with decision_function
        y_test_proba = model.decision_function(X_test)
    
    # Store metrics
    metrics.append({
        'Model': name,
        'Train Accuracy': accuracy_score(y_train, y_train_pred),
        'Test Accuracy': accuracy_score(y_test, y_test_pred),
        'Train Precision': precision_score(y_train, y_train_pred, average='weighted'),
        'Test Precision': precision_score(y_test, y_test_pred, average='weighted'),
        'Train Recall': recall_score(y_train, y_train_pred, average='weighted'),
        'Test Recall': recall_score(y_test, y_test_pred, average='weighted'),
        'Train F1': f1_score(y_train, y_train_pred, average='weighted'),
        'Test F1': f1_score(y_test, y_test_pred, average='weighted'),
        'Test ROC-AUC': roc_auc_score(y_test, y_test_proba)
    })

# Convert to DataFrame
metrics_df = pd.DataFrame(metrics)
metrics_df
```
